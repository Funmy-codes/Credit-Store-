```python
## I noticed that there was some error in the seperation of the categorical data from the numerical data
## I decided to clean the data set
# Age
df['Age'].sample(200).values
df.Age = df['Age'].str.replace('-', '')
df.Age = df['Age'].str.replace('_', '')
df.Age = df.Age.astype(int)
df.Age.dtypes
```

```python
# Num_of_Loan
df['Num_of_Loan'].sample(200).values
df.Num_of_Loan = df['Num_of_Loan'].str.replace('-', '')
df.Num_of_Loan = df['Num_of_Loan'].str.replace('_', '')
df.Num_of_loan = df['Num_of_Loan'].astype(str).str.replace('.', '')
df.Num_of_Loan = df.Num_of_Loan.astype(float)
df.Num_of_Loan.dtypes
```

```python
# Num_of_Delayed_Payment
df['Num_of_Delayed_Payment'].sample(200).values
df.Num_of_Delayed_Payment = df['Num_of_Delayed_Payment'].str.replace('_', '')
df.Num_of_Delayed_Payment = df.Num_of_Delayed_Payment.astype(float)
df.Num_of_Delayed_Payment.dtypes
```

```python
# Changed_Credit_Limit
df['Changed_Credit_Limit'].sample(200).values
df.Changed_Credit_Limit = df['Changed_Credit_Limit'].str.replace('-', '')
df.Changed_Credit_Limit = df['Changed_Credit_Limit'].str.replace('_', '')
df.Changed_Credit_Limit = pd.to_numeric(df.Changed_Credit_Limit, errors = 'coerce')
df.Changed_Credit_Limit.dtypes
```

```python
# Outstanding_Debt
df['Outstanding_Debt'].sample(200).values
df.Outstanding_Debt = df['Outstanding_Debt'].str.replace('-', '')
df.Outstanding_Debt = df['Outstanding_Debt'].str.replace('_', '')
df.Outstanding_Debt = df.Outstanding_Debt.astype(float)
df.Outstanding_Debt.dtypes
```

```python
# Amount_invested_monthly
df['Amount_invested_monthly'].sample(200).values
df.Amount_invested_monthly = df['Amount_invested_monthly'].str.replace('-', '')
df.Amount_invested_monthly = df['Amount_invested_monthly'].str.replace('_', '')
df.Amount_invested_monthly = df.Amount_invested_monthly.astype(float)
df.Amount_invested_monthly.dtypes
```

```python
# Monthly_Balance
df['Monthly_Balance'].sample(200).values
df.Monthly_Balance = df['Monthly_Balance'].str.replace('-', '')
df.Monthly_Balance = df['Monthly_Balance'].str.replace('_', '')
df.Monthly_Balance = df.Monthly_Balance.astype(float)
df.Monthly_Balance.dtypes
```

```python
## I confirmed here if my categorical and numerical data are well arranged
categorical = df.select_dtypes(include = ['object', 'category'])
numerical = df.select_dtypes(include = 'number')
print(f'Categorical df')
display(categorical.head(2))
print(f'\n\nNumerical df')
display(numerical.head(2))
```

```python
# I worked on the columns that has null values
for i in df.columns:
  if((df[i].isnull().sum()/ len(df)) * 100) > 30:
    df.drop([i], axis = 1, inplace = True)

  elif df[i].dtypes == 'O':
    df[i].fillna(df[i].mode()[0], inplace = True)

  else:
    df[i].fillna(df[i].median(), inplace = True)

df.isnull().sum()
```

```python
# I checked to verify
df.isnull().sum()
```

```python
df.describe()
```

```python
## I checked for outliners
def outlierRemoval(dataframe):
  for i in dataframe.columns:
    if dataframe[i].dtypes != 'O':
      Q1 = dataframe[i].describe()[4]
      Q3 = dataframe[i].describe()[6]
      IQR = Q3 - Q1
      minThreshold = Q1 - 1.5 * IQR
      maxThreshold = Q3 + 1.5 * IQR

      dataframe = dataframe.loc[(dataframe[i] >= minThreshold) & (dataframe[i] <= maxThreshold)]
  return dataframe
```

```python
df = outlierRemoval(df)
df
```
